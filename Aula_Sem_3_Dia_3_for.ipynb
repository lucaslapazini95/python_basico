```python
#escrever na tela os número de 0 a 9
contador = 1 #inicia o contador
while contador <= 10: #condição de parada
  print (contador)
  contador += 1 #incrementa o contador
```

```python
# for i in range(inicio_do_contador, condição de parada, incremento ou decremento)
#no caso abaixo, ele toma algumas coisas como padrão
#contador = 0
#incremento aumenta de 1 em 1

for contador in range(10):
  print(contador)
```

```python
for contador in range(1, 11):
  print(contador)
```

```python
for contador in range (5, 20, 3):
  print(contador)
```

```python
fruta = 'Banana'

for contador in fruta:
  print(contador)
```

```python
#Escrever na tela os número de 1 a 20 em ordem decrescente
for i in range(20,0,-1):
  print(i, end=' ')
```

Se preciso de **validação**, tenho que **obrigar algo a ser daquele jeito**, uso o **while**.
**Repetições** usamos o **for**, como percorrer colunas de dados, **casos que sabemos a quantidade de vezes que vai executar a repetição**.

```python
#1) Crie um programa em Python que escreva
#na tela os números de 250 a 120 em ordem decrescente.

for i in range(250, 119, -1):
  print(i)
```

```python
#2) Crie um programa em Python que escreva na tela os
#números de 1000 a 1500 em ordem crescente

for i in range(1000, 1501):
  print(i)
```
